```python
%load_ext autoreload
%autoreload 2

import os,sys,inspect
current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parent_dir = os.path.dirname(current_dir)
parent_parent_dir = os.path.dirname(current_dir)

sys.path.insert(0, os.path.join(parent_dir)) 
sys.path.insert(0, os.path.join(parent_parent_dir)) 

import numpy as np
from topo2vec.common.geographic.geo_map import GeoMap
import topo2vec.modules.topography_profiler as tp
import topo2vec.modules.visual_topography_profiler as vtp
from shapely.geometry import Point, Polygon
from topo2vec.common.visualizations import convert_multi_radius_ndarray_to_printable

import folium
```

```python
polygon = Polygon([Point(5, 45), Point(5, 45.1), Point(5.1, 45.1),
                   Point(5.1, 45.1), Point(5, 45)])
points_inside = [Point(5.0658811, 45.0851164),
              Point(5.058811, 45.01164)]

```

```python
the_map = vtp.TopoMap(polygon)
the_map.add_random_class_points(polygon, 100, 'peaks', 'red')
the_map.add_random_class_points(polygon, 100, 'streams', 'blue')
the_map.add_random_class_points(polygon, 100, 'saddles', 'green')
the_map.add_random_class_points(polygon, 100, 'cliffs', 'orange')
the_map.get_folium_map()
```

```python
the_map = vtp.TopoMap(polygon)
the_map.add_similar_points(points_inside, polygon, 500, n=10, color='red')
the_map.get_folium_map()
```

```python
NUM_TO_PICK = 5

the_map = vtp.TopoMap(polygon)
np_points_used, np_points_patches_used = tp.get_all_class_points_in_polygon(polygon, 500, 'peaks')
picked_indices = np.random.choice(list(range(len(np_points_used))), NUM_TO_PICK)

points_used_picked_np = np_points_used[picked_indices]
points_used_picked = [Point(x[0],x[1]) for x in points_used_picked_np]
points_patches_used_picked = np_points_patches_used[picked_indices]
points_patches_used_picked = convert_multi_radius_ndarray_to_printable(points_patches_used_picked, dir=False)

the_map.add_similar_points(points_used_picked, polygon, 500, n=10, color='red')
the_map.add_points_with_images(points_used_picked_np, points_patches_used_picked, color='green')
the_map.get_folium_map()
```
